```python
# Дана функция f(x) = 18x^3+5x^2 + 10x - 30
# 1 Определить корни
# 2 Найти интервалы, на которых функция возрастает
# 3 Найти интервалы, на которых функция убывает
# 4 Построить график
# 5 Вычислить вершину
# 6 Определить промежутки, на котором f > 0
# 7 Определить промежутки, на котором f < 0
import sympy
# import matplotlib
a = 18
b = 5
c = 10
d = -30
```

```python
# 1 Определить корни
x = sympy.Symbol('x')
y = sympy.solve(a*x**3 + b*x**2 + c*x + d, x)
print(y)
```

```python
# 2 Найти интервалы, на которых функция возрастает  
# --> ВОЗРАСТАЮЩАЯ ФУНКЦИЯ => экстремума (точки максимума и минимума) нет
print(sympy.solve(a*x**3 + b*x**2 + c*x + d > 0, x))
```

```python
# 3 Найти интервалы, на которых функция убывает 
# => ВОЗРАСТАЮЩАЯ ФУНКЦИЯ
print(sympy.solve(a*x**3 + b*x**2 + c*x + d < 0, x))
```

```python
# 4 Построить график
sympy.plotting.plot(a*x**3 + b*x**2 + c*x + d, x)
```

```python
# 5 Вычислить вершину
# ФУНКЦИЯ ВОЗРАСТАЮЩАЯ И ВЕРШИНЫ НЕТ !
f = a*x**3 + b*x**2 + c*x + d
dif = sympy.diff(f, x)
print(dif)
print(sympy.solve(dif, x))
```

```python
# 6 Определить промежутки, на котором f > 0
print(sympy.solve(f > 0, x))
```

```python
# 7 Определить промежутки, на котором f < 0
print(sympy.solve(f < 0, x))
```
